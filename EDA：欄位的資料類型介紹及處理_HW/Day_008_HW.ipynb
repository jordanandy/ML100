```python
import statistics

print(app_train["AMT_INCOME_TOTAL"].mean())
print(app_train["AMT_INCOME_TOTAL"].std())


mean = statistics.mean(app_train["AMT_INCOME_TOTAL"])
stdev = statistics.stdev(app_train["AMT_INCOME_TOTAL"])



print(mean)
print(stdev)


```
